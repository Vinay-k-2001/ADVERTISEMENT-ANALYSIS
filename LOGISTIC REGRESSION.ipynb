```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
# 0 - not purchased ,  1 - purchased
# it is categorical dataset (i.e., dataset referred to classification)
df = pd.read_csv('/content/Social_Network_Ads.csv')
df
```

```python
print(df.info())
print('\n')
print(df.shape)
print('\n')
print(df.size)
```

```python
# used to select the part of a dataframe
# integer location based indexing -iloc
# iloc[[a:b],[c;d]] 
# a,b = start,end of rows ; c,d = start,end of columns

# two dimensional input data 
x = df.iloc[:,[2,3]].values 

# one dimensional output 
y = df.iloc[:,-1].values    
```

```python
x
```

```python
y
```

**TRAINING AND TESTING using SKLEARN**

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=0,test_size=0.25)
```

```python
# default splitting = 75:25
print(x_train.shape)
print(x_test.shape)
```

**NORMALISATION**

```python
# NORMALIZATION STEP 
# SCALING THE VALUES
```

```python
#EXAMPLE FOR USING MIN MAX SCALER(numpy implementation)
# AGE = 18-60 , SALARY = 15k - 150k

# GENEREAL VISUALISATION

np.random.seed(100)
x1 = np.random.randint(18,60,30)
x1 = np.sort(x1)
x2 = np.random.randint(15000,150000,30)
plt.plot(x1,x2)
plt.show()
```

```python
# VISUALISATION BY USING NORMALISATION
# ALL VALUES in b/n 0 & 1

x1min = min(x1)
x1max = max(x1)
x2min = min(x2)
x2max = max(x2) 

x1norm = (x1 - x1min)/(x1max - x1min)

x2norm = (x2 - x2min)/(x2max - x2min)

plt.plot(x1norm,x2norm)
```

```python
# scaling is done for inputs only (x_train - fit_transform, x_test - transform)
# MINMAXSCALER :
'''
x1norm = (x1 - x1min)/(x1max - x1min)

x2norm = (x2 - x2min)/(x2max - x2min)
'''

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
# now all the values are in between 0 and 1

x_train 
```

```python
# IMPORTING LOGISTIC REGESSION

from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
# FITTING THE MODEL

model.fit(x_train,y_train)
```

```python
# PREDICTING

y_pred = model.predict(x_test)
y_pred
```

```python
y_test
```

```python
# actual output 

np.unique(y_test,return_counts=True) 
```

```python
# predicted output 

np.unique(y_pred,return_counts=True) 
```

**ACCURACY SCORE**

```python
# ACCURACY SCORE FOR ALL CLASIFIERS

from sklearn.metrics import accuracy_score,confusion_matrix 
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```
